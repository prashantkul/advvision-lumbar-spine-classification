```python
import pandas as pd
import vision_models.constants as constants
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# labels_df = pd.read_csv(constants.TRAIN_LABEL_PATH)
# labels_df = pd.read_csv("/opt/dataset/train.csv")
labels_df = pd.read_csv("/opt/dataset/train_label_coordinates.csv")
labels_df
# print(labels_df.groupby(['condition', 'level'])['study_id','series_id'].nunique())

# Count unique rows per condition and level
unique_counts = labels_df.groupby(['condition', 'level']).nunique()[['study_id', 'series_id']]

# Plotting
plt.figure(figsize=(12, 8))
sns.heatmap(unique_counts, annot=True, fmt='d', cmap='viridis')
plt.title('Unique Count of Study and Series IDs per Condition and Level')
plt.xlabel('Unique Counts')
plt.ylabel('Condition and Level')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
```
