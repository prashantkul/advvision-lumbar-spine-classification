```python
import pandas as pd
import vision_models.constants as constants

# Load the predictions CSV file
predictions_df = pd.read_csv('data_visual_outputs/predictions.csv')
predictions_df.head()

# Define the threshold
threshold = constants.DISEASE_THRESHOLD

# Convert the probabilities to binary values
binary_predictions_df = (predictions_df.iloc[:, 2:] > threshold).astype(int)

# Add back the study_id and series_id columns
binary_predictions_df.insert(0, 'series_id', predictions_df['series_id'])
binary_predictions_df.insert(0, 'study_id', predictions_df['study_id'])

print(binary_predictions_df.head())

# Save the binary results to a new CSV file
binary_predictions_df.to_csv('data_visual_outputs/binary_predictions.csv', index=False)
print("Binary predictions saved to binary_predictions.csv")

# Aggregate predictions by study_id and take the max for each condition
aggregated_predictions_df = binary_predictions_df.groupby('study_id').max().drop(columns = {'series_id'}).reset_index()

print(aggregated_predictions_df.head())

# Save the aggregated binary results to a new CSV file
aggregated_predictions_df.to_csv('data_visual_outputs/aggregated_binary_predictions.csv', index=False)
print("Aggregated binary predictions saved to aggregated_binary_predictions.csv")
```
